```python
import pandas as pd
import numpy as np
import warnings

warnings.filterwarnings("ignore")
```

```python
data = pd.read_excel("./data/predict_population.xlsx")
data.columns = ['성별','시점','연령별','인구수']
data["총인구수"] = data.groupby(['성별','시점'])['인구수'].transform(sum)
data['인구비율'] = np.round(data['인구수'] / data['총인구수'] * 100,4)
data
```

```python
data.to_excel('./output/인구예측수.xlsx',index=False)
```
