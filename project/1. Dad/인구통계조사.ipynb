```python
import pandas as pd
import numpy as np
import warnings

warnings.filterwarnings("ignore")
```

```python
data = pd.read_excel("./data/population.xlsx")
data.columns = ['sido','age','year','cnt']
data = data[["sido","age","cnt"]]
data["age"] = data['age'].str.replace("세","").astype(int)
data
```

```python
total = data.groupby(['age'])['cnt'].agg(sum).reset_index()
total["sido"] = '전국'
total = total[["sido","age","cnt"]]
```

```python
df = pd.concat([total,data],axis = 0)
df
```

```python
total_population = total.groupby(['sido'])['cnt'].agg(sum).values[0]
```

```python
year2024 = total.copy()
year2024["cnt2"] = year2024['cnt'].shift(1).fillna(0).astype(int)
year2024 = year2024[["sido","age","cnt2"]]
year2024['age_group'] = pd.cut(
    data['age'],
    bins=range(0, data['age'].max() + 6, 5),
    right=False,
    labels=[f"{i}~{i+4}" for i in range(0, data['age'].max() + 1, 5)]
)

year2024_result = year2024.groupby('age_group')['cnt2'].sum().reset_index()
```
